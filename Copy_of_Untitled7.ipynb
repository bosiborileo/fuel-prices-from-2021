```python
import pandas as pd

# Read the XLSX files into DataFrames

df1 = pd.read_excel('/content/drive/MyDrive/epra fuel/14th-February-2023-–-15th-March-2023.xlsx')
df4 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-December-2022-14th-January-2023.xlsx')
df5 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-February-14th-March-2022_Website.xlsx')
df6 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-January-14th-February-2022.xlsx')
df7 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-January-2023-–14th-February-2023-11.xlsx')
df8 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-March-2022-to-April-2022.xlsx')
df9 = pd.read_excel('/content/drive/MyDrive/epra fuel/FINAL-ACTUAL-PUMP-PRICES_15th-MARCH-14th-APRIL-2023.xlsx')
df11 = pd.read_excel('/content/drive/MyDrive/epra fuel/Prices-15th-Nov-14th-Dec-2022 (1).xlsx')
df12 = pd.read_excel('/content/drive/MyDrive/epra fuel/Prices-August-to-September-2022 (1).xlsx')
df13 = pd.read_excel('/content/drive/MyDrive/epra fuel/Prices-Oct-to-Nov-2022 (1).xlsx')
df14 = pd.read_excel('/content/drive/MyDrive/epra fuel/Prices-july-to-August-2022 (1).xlsx')
df15 = pd.read_excel('/content/drive/MyDrive/epra fuel/Prices-september-to-october-2022 (1).xlsx')
df16 = pd.read_excel('/content/drive/MyDrive/epra fuel/Pump-Prices-15-July-14-August-2023-Website.xlsx')
df17 = pd.read_excel('/content/drive/MyDrive/epra fuel/Pump-Prices-15-June-14-July-2022_Website.xlsx')
df18 = pd.read_excel('/content/drive/MyDrive/epra fuel/Pump-Prices11-15-May-14-June-2022_Website.xlsx')
df19 = pd.read_excel('/content/drive/MyDrive/epra fuel/15th-January-2023-–14th-February-2023-11.xlsx')


# Combine the DataFrames along the row axis, ignoring the index
combined_xlsx = pd.concat([df1, df4, df5, df6, df7, df8, df9, df11, df12, df13, df14, df15, df16, df17, df18, df19], ignore_index=True)

# Save the combined DataFrame to a new XLSX file
combined_xlsx.to_excel('combined_new.xlsx', index=False)

import os
# Get a list of all the XLSX files in the directory
xlsx_files = [f for f in os.listdir() if f.endswith('.xlsx')]

# Read the XLSX file into a Pandas DataFrame
df = pd.read_excel('combined_new.xlsx')

# Save the DataFrame to a CSV file
df.to_csv('file.csv', index=False)

#combining the csv files
df22 = pd.read_csv('file.csv')
df2 = pd.read_csv('/content/drive/MyDrive/epra fuel/15th-April-2022-14th-May-2022.csv')
df3 = pd.read_csv('/content/drive/MyDrive/epra fuel/15th-April-2023-to-14th-may-2023 (1).csv')
df21 = pd.read_csv('/content/drive/MyDrive/epra fuel/15th-November-14th-December-2021-1 (1).csv')
df10 = pd.read_csv('/content/drive/MyDrive/epra fuel/July-1-July-14-2023 (1).csv')
df20 = pd.read_csv('/content/drive/MyDrive/epra fuel/15th-April-2023-to-14th-may-2023.csv')

df_combined = pd.concat([df22, df2,df3, df21, df10, df20], axis=1)
df_combined.to_csv('combined.csv', index=False, encoding='utf-8')
```
